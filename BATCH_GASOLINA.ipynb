```python
#Comandos usados pra concatenar os dataframes de gasolina : 

d2161 = pd.read_csv("/content/G2016-01.csv")
d2162 = pd.read_csv("/content/G2016-02.csv")
d2171 = pd.read_csv("/content/G2017-01.csv")
d2172 = pd.read_csv("/content/G2017-02.csv")
d2181 = pd.read_csv("/content/G2018-01.csv")
d2182 = pd.read_csv("/content/G2018-02.csv")
d2191 = pd.read_csv("/content/G2019-01.csv")
d2192 = pd.read_csv("/content/G2019-02.csv")
d2201 = pd.read_csv("/content/G2020-01.csv")
d2202 = pd.read_csv("/content/G2020-02.csv")
d2211 = pd.read_csv("/content/G2021-01.csv")
d2212 = pd.read_csv("/content/G2021-02.csv")

df = pd.concat([d2161, d2162, d2171, d2172, d2181, d2182,
          d2191, d2192, d2201, d2202, d2211 ], axis=0, ignore_index=True, 
          verify_integrity=True, join="inner")

df.to_csv("Gasolina_Concatenado.csv")
```
